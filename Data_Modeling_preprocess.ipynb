```python
import pandas as pd
import numpy as np
```

## Data Preprocessing

```python
data = pd.read_csv("preprocessed_dataset/new_trade_df.csv")
trades_df = pd.DataFrame(data)
trades_df
```

```python
nft_df_1 = pd.read_csv("preprocessed_dataset/ntf_df1.csv")
nft_df_2 = pd.read_csv("preprocessed_dataset/ntf_df2.csv")
nft_df_3 = pd.read_csv("preprocessed_dataset/ntf_df3.csv")
nft_df_4 = pd.read_csv("preprocessed_dataset/ntf_df4.csv")
nft_df_5 = pd.read_csv("preprocessed_dataset/ntf_df5.csv")
nft_df = pd.concat([nft_df_1, nft_df_2, nft_df_3, nft_df_4, nft_df_5])
```

```python
nft_df
```

```python
trades_df = trades_df[["token_ids","block_timestamp","price_usd"]]
trades_df
```

```python
nft_df=nft_df.rename(columns={"token_id":"token_ids"})
nft_df
```

```python
nft_df = nft_df[['token_ids', 'Artifact', 'Category',
       'Eastern Resource', 'Eastern Resource Tier', 'Environment',
       'Environment Tier', 'Koda', 'Northern Resource',
       'Northern Resource Tier', 'Obelisk Piece', 'Plot', 'Sediment',
       'Sediment Tier', 'Southern Resource', 'Southern Resource Tier',
       'Western Resource', 'Western Resource Tier']]
```

```python
df = nft_df.merge(trades_df, on='token_ids', how='left')
df
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
df["token_ids"]=le.fit_transform(df["token_ids"])
```

```python
df
```

```python
df["Artifact"]=le.fit_transform(df["Artifact"])
```

```python
df["Category"]=le.fit_transform(df["Category"])
df["Eastern Resource"]=le.fit_transform(df["Eastern Resource"])
df["Environment"]=le.fit_transform(df["Environment"])
df["Koda"]=le.fit_transform(df["Koda"])
df["Northern Resource"]=le.fit_transform(df["Northern Resource"])
df["Obelisk Piece"]=le.fit_transform(df["Obelisk Piece"])
df["Sediment"]=le.fit_transform(df["Sediment"])
df["Southern Resource"]=le.fit_transform(df["Southern Resource"])
df["Western Resource"]=le.fit_transform(df["Western Resource"])
```

```python
df
```

```python
df["block_timestamp"]=df["block_timestamp"].fillna("2022-10-18")
df=df.fillna(0.0)
```

```python
df
```

```python
df["Eastern Resource Tier"]=le.fit_transform(df["Eastern Resource Tier"])
df["Environment Tier"]=le.fit_transform(df["Environment Tier"])
df["Northern Resource Tier"]=le.fit_transform(df["Northern Resource Tier"])
df["Sediment Tier"]=le.fit_transform(df["Sediment Tier"])
df["Southern Resource Tier"]=le.fit_transform(df["Southern Resource Tier"])
df["Western Resource Tier"]=le.fit_transform(df["Western Resource Tier"])
```

```python
df
```

```python
temp=max(df["Koda"])
df["koda_or_not"]=[1 if i!=temp else 0 for i in df["Koda"]]
```

```python
df
```

```python
from datetime import datetime
```

```python
df["block_timestamp"]=[datetime.strptime(i,"%Y-%m-%d").timestamp() for i in df["block_timestamp"]]
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
df["Plot"]=ss.fit_transform(np.array(df["Plot"]).reshape(-1, 1))
```

```python
df
```

```python
df["block_timestamp"]=ss.fit_transform(np.array(df["block_timestamp"]).reshape(-1, 1))
```

```python
df
```

```python
df_test=df[df["price_usd"]==0]
df_test=df_test.drop(columns=['price_usd'])
```

```python
df_test
```

```python
df_train=df[df["price_usd"]!=0]
```

```python
df_train
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_train.drop(columns=["price_usd","token_ids"])
y=df_train["price_usd"]
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, random_state=42)
```
