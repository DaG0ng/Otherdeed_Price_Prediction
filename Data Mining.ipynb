```python
import requests
result=[]
url_set=[]
for i in range(100000):
    url = "https://deep-index.moralis.io/api/v2/nft/0x34d85c9CDeB23FA97cb08333b511ac86E1C4E258/"+str(i)+"?chain=eth&format=decimal"
    url_set.append(url)
```

```python
for i in tqdm(range(25000)):   
    headers = {
        "accept": "application/json",
        "X-API-Key": "1Qpz7ycsCNu9YB5EByIagkkOn2x1CmJ2I7JfWeu3asNBeilXHrBHFUvfcu9xts5e"
    }
    response = requests.get(url_set[i], headers=headers)
    result.append(response.text)
```

```python
import json

with open('data1.json', 'w') as f:
    json.dump(result, f)
```
