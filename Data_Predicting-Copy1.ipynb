## Machine Learning Algorithms

## Gradient Boosting XGBoost

```python
import xgboost
from sklearn.model_selection import GridSearchCV 
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
model = xgboost.XGBRegressor(n_estimators=500, max_depth=8, eta=0.05, subsample=0.9, colsample_bytree=0.9)
```

```python
model.fit(X,y)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_valid,y_valid)
```

```python
mean_absolute_error(y_valid,model.predict(X_valid))
```

```python
predict_precentage_diff=abs(y_valid-model.predict(X_valid))/y_valid*100
```

```python
show_dic={}
show_dic["0-20% diff"]=0
show_dic["20-40% diff"]=0
show_dic["40-60% diff"]=0
show_dic["60-80% diff"]=0
show_dic["80-100% diff"]=0
show_dic["more than 100% diff"]=0
```

```python
for i in np.array(predict_precentage_diff):
    if i >=0 and i<=20:
        show_dic["0-20% diff"]+=1
    elif i >=20 and i<=40:
        show_dic["20-40% diff"]+=1
    elif i >=40 and i<=60:
        show_dic["40-60% diff"]+=1
    elif i >=60 and i<=80:
        show_dic["60-80% diff"]+=1
    elif i >=80 and i<=100:
        show_dic["80-100% diff"]+=1
    else:
        show_dic["more than 100% diff"]+=1
```
