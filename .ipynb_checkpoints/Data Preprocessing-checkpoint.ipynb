```python
import json
import pandas as pd
import numpy as np
import plotly
```

```python
import tqdm
```

```python
json_file_path = "./raw_dataset/data1.json"
json_file_path2 = "./raw_dataset/data2.json"
json_file_path3 = "./raw_dataset/data3.json"
json_file_path4 = "./raw_dataset/data4.json"
```

```python
with open(json_file_path, 'r') as j:
    contents = json.loads(j.read())
with open(json_file_path2, 'r') as j:
    contents2 = json.loads(j.read())
with open(json_file_path3, 'r') as j:
    contents3 = json.loads(j.read())
with open(json_file_path4, 'r') as j:
    contents4 = json.loads(j.read())
```

```python
content = contents[:25000]+contents2+contents3+contents4
```

```python
len(content)
```

```python
for i in range(len(content)):
    content[i] = json.loads(content[i])
```

```python
content=[i for i in content if "message" not in i.keys()]
```

```python
len(content)
```

```python
for i in range(len(content)):
    content[i]["metadata"] = json.loads(content[i]["metadata"])
```

```python
content[0]
```

```python
attributes_list=[]
for i in content:
    for j in i["metadata"]["attributes"]:
        attributes_list.append(j["trait_type"])
```

```python
len(attributes_list)
```

```python
attributes_list=list(np.unique(attributes_list))
```

```python
attributes_list
```

```python
dfItem = pd.DataFrame.from_records(content)
```

```python
dfItem=dfItem.set_index("token_id")
```

```python
dfItem
```

```python
attribute_dic={}
for i in attributes_list:
    attribute_dic[i]=[]
```

```python
nft_metadata_lst=[]
for _,j in dfItem.iterrows():
    temp_list=j["metadata"]["attributes"]
    temp_dic={}
    for i in temp_list:
        temp_dic[i["trait_type"]]=i["value"]
    nft_metadata_lst.append(temp_dic)
```

```python
len(nft_metadata_lst)
```

```python
for i in nft_metadata_lst:
    for j in attribute_dic.keys():
        try:
            attribute_dic[j].append(i[j])
        except:
            attribute_dic[j].append(None)
```

```python
for i in attribute_dic.keys():
    dfItem[i]=attribute_dic[i]
```

```python
dfItem
```

```python
dfItem.iloc[0:20000].to_csv("./preprocessed_dataset/ntf_df1.csv",index=False)
dfItem.iloc[20000:40000].to_csv("./preprocessed_dataset/ntf_df2.csv",index=False)
dfItem.iloc[40000:60000].to_csv("./preprocessed_dataset/ntf_df3.csv",index=False)
dfItem.iloc[60000:80000].to_csv("./preprocessed_dataset/ntf_df4.csv",index=False)
dfItem.iloc[80000:99939].to_csv("./preprocessed_dataset/ntf_df5.csv",index=False)
```
